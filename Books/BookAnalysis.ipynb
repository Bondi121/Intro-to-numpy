## Are books with fewer pages rated higher than those with large page counts?
There isn't a large difference between the book ratings. 
Only 0.06 between the top book in each category.

This may benefit from a chart to produce a better visual.

```python
few_pages = books[books['num_pages'] < 300]
few_pages = few_pages[few_pages['ratings_count'] > 1000]
few_pages.sort_values(by=['average_rating'], ascending=False) [:1]
```

```python
many_pages = books[books['num_pages'] > 300]
many_pages = many_pages[many_pages['ratings_count'] > 1000]
many_pages.sort_values(by=['average_rating'], ascending=False) [:1]
```

## Who wrote the most pages?

```python
max = 0
top_author = None

all_authors = books['authors'].unique()

for author in all_authors:
    total_pages = books.loc[books['authors'] == author, 'num_pages'].sum()
    if total_pages > max:
        max = total_pages
        top_author = author 
        print(max)
        print(author)
        
print(max)
print(top_author)
```

## What's an author's average page count?

```python
their_pages = books.loc[books['authors'] == 'Stephen King', 'num_pages'].sum()
their_books = (books['authors'] == 'Stephen King').value_counts() [1]
avg_pages = their_pages / their_books
print(avg_pages)
```
