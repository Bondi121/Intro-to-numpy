## What is the average number of pets found in a month in the year 2015? Are there months where there is a higher number of animals found?

```python
average_number_month = intakes[intakes['datetime'] > "2015-01-01T00:0:00.000"]
average_number_month  = average_number_month [average_number_month ['datetime2'] < "2015-02-01T00:0:00.000"]
average_number_month.count()
```

```python
average_number_month = intakes[intakes['datetime'] > "2015-01-01T00:0:00.000"]
average_number_month  = average_number_month [average_number_month ['datetime2'] < "2016-01-01T00:0:00.000"]
average_number_month.count()
```

```python
average_number_month.count() / 12
```

## What is the ratio of incoming pets vs. adopted pets?

```python
intakes['animal_id'].count()
```

```python
adaptions = outcomes[outcomes['outcome_type'] == "Adoption"]
adaptions.count()
```

```python
ratio = intakes['animal_id'].count() / adaptions.count()  
ratio
```

## What are the adoption rates for specific breeds?

```python
grouped_by_breeds = intakes.groupby('breed')
grouped_by_breeds.size().sort_values(ascending=False)[:5]
```

```python
grouped_by_breeds.size().sort_values(ascending=False)[:5]
```

```python
adoptions = outcomes[outcomes['outcome_type'] == "Adoption"]
adoptions = adoptions[adoptions['breed'] == "Domestic Shorthair Mix"]
adoptions.count()
```

```python
domestic_shorthair = intakes[intakes['breed'] == "Domestic Shorthair Mix"]
domestic_shorthair.count() / adoptions.count()
```
