## North American Sales by Genre: Create a pie chart to show the percentage of sales each genre had in the North American market. (Group smaller “slices” together into an “other” category to create an easier to read chart) Detail what is included in the other category. Explain the chart using markdown.

```python
NA_sales_grouped_by_genre = videogames[['Genre', 'NA_Sales']].groupby('Genre').sum()
NA_sales_grouped_by_genre= NA_sales_grouped_by_genre.sort_values('NA_Sales')
NA_sales_grouped_by_genre
```

```python
Shooter = NA_sales_by_genre.loc['Shooter'].values[0]
Action = NA_sales_by_genre.loc['Action'].values[0]
Platform = NA_sales_by_genre.loc['Platform'].values[0]
Sports = NA_sales_by_genre.loc['Sports'].values[0]
Other = (NA_sales_by_genre.loc['Misc'].values[0] +
      NA_sales_by_genre.loc['Fighting'].values[0] +
      NA_sales_by_genre.loc['Puzzle'].values[0] +
      NA_sales_by_genre.loc['Simulation'].values[0] +
      NA_sales_by_genre.loc['Role-Playing'].values[0] +
      NA_sales_by_genre.loc['Racing'].values[0])

index = ['Other', 'Shooter', 'Action', 'Platform', 'Sports']
colors = ( "orange", "cyan", "brown",
          "grey", "beige")

explode = (0, 0, 0, 0, 0.1)

wp = { 'linewidth' : 2, 'edgecolor' : "black" }

NA_sales_by_genre_df = pd.DataFrame({'NA_Sales': [Shooter, Action, Platform, Sports, Other]
                  }, index=index)

plot = NA_sales_by_genre_df.plot.pie(y = 'NA_Sales', colors=colors, startangle = 90, wedgeprops = wp, explode=explode, autopct="%.1f", figsize=(11, 8))
plt.title("Genre divided by the North American Market in Sales (Percentage)")
```
