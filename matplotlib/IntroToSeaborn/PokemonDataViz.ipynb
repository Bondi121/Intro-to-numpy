```python
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme()
```

```python
pokemon = pd.read_csv('pokemon_40.csv')
pokemon.head()
```

What is the relationship between Attack and HP?

```python
sns.scatterplot(data=pokemon, x='HP', y='Attack', hue='Type')
```

```python
sns.lineplot(data=pokemon, x='HP', y='Attack')
```

```python
sns.lineplot(data=pokemon, x='Defense', y='Attack')
```

```python
sns.scatterplot(data=pokemon, x='HP', y='Attack', hue='Type')
```

```python
sns.relplot(data=pokemon, x='HP', y='Attack', hue='Type', col='Type')
```

```python
sns.relplot(data=pokemon, x='HP', y='Attack', hue='Type', col='Type', col_wrap=3)
```

Generally, HP and Attack are somewhat positively correlated. Pokemon with more HP tend to be stronger. 

## What is the distribution of Attack?

```python
sns.histplot(data=pokemon, x='Attack', kde=True)
```

```python
sns.histplot(data=pokemon, x='Attack', bins=15)
```

```python
sns.kdeplot(data=pokemon, x='Attack')
```

```python
sns.histplot(data=pokemon, x='Attack', bins=10, kde=True)
```

```python
sns.displot(data=pokemon, x='Attack', bins=10, col="Type", col_wrap=3)
```

Most Pokemon have an Attack distributed between 50 and 120.

### What is the relationship between Attack and Type? (Categorical Scatter Plot)

```python
sns.stripplot(data=pokemon, x='Type', y='Attack', hue='HP')
```

```python
sns.catplot(kind='strip', data=pokemon, x='Type', y='Attack', aspect=2, hue='HP')
```

### What is the distribution of Attack of each Type? (Categorical Distribution Plots)

```python
sns.catplot(kind='box', data=pokemon, x='Type', y='Attack', aspect=2)
```

```python
sns.catplot(kind='violin', data=pokemon, x='Type', y='Attack', aspect=2)
```

```python
sns.catplot(kind='boxen', data=pokemon, x='Type', y='Attack', aspect=2)
```

For the Normal Type Pokemon, the minimum Attack is 5, the maximum is 110, the median is 75. This first and third quartiles are 55, and 105, respectively.
