```python
import pandas as pd
import seaborn as sns

sns.set_theme()

monsters = pd.read_csv('monsters_40.csv')

monsters.head()
```

## Challenge 2

What is the relationship between:
- Level and attack points?
- Level and Defense Points?
Bonus:use sns.relplot(kind="scatter") to solve one of the these questions.

```python
sns.scatterplot(data=monsters, x='Level', y='Attack_Points', hue='Attributes')
```

```python
sns.scatterplot(data=monsters, x='Level', y='Attack_Points', hue='Attributes')
```

```python
sns.scatterplot(data=monsters, x='Level', y='Defense_Points', hue='Attributes')
```

```python
sns.relplot(data=monsters, x='Level', y='Defense_Points', kind="scatter")
```

```python
sns.relplot(data=monsters, x='Level', y='Defense_Points', kind="line")
```
